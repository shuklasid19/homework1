* Is it possible that an event is 
independent of itself? If so, when?
A:  Let A be an event. If A is independent of itself, then P(A) = P(A\A) = P(A)2, so P(A) is 0 or 1. So this is only possible in the extreme cases that the event has probability 0 or 1.
